#### **But DU MODELE:**
#### **Prédire et prévenir les blessures causées lors de matchs de tennis.**

```python
import pandas as pd
```

```python
# Charger les données
df_blessures = pd.read_excel('../DataSets/BlessuresTennis_Causes.xlsx')
df_blessures.head()
```

```python
# Vérifier les valeurs manquantes
print("Valeurs manquantes dans les données de blessures :")
df_blessures.isnull().sum()
```

```python
df_joueurs = pd.read_csv('../../Nettoyage_data/Datasets/Joueurs/KagglePlayers_nettoye.csv')
df_joueurs.head()
```

```python
print("\nValeurs manquantes dans les données des joueurs :")
df_joueurs.isnull().sum()
```

```python
df_matchs = pd.read_csv('../../Nettoyage_data/Datasets/Matchs/KaggleMatches_cleaned.csv')
df_matchs.head()
```

```python
print("\nValeurs manquantes dans les données des matchs :")
df_matchs.isnull().sum()
```

```python
df_terrains = pd.read_csv('../../Nettoyage_data/Datasets/Terrains/tennis_court_nettoye.csv')
df_terrains.head()
```

```python
print("\nValeurs manquantes dans les données des terrains :")
df_terrains.isnull().sum()
```

```python
# Supprimer les doublons s'il y en a
df_blessures.drop_duplicates(inplace=True)
df_joueurs.drop_duplicates(inplace=True)
df_matchs.drop_duplicates(inplace=True)
df_terrains.drop_duplicates(inplace=True)
```

```python
# Fusionner les données des blessures avec les joueurs
df_blessures_joueurs = pd.merge(df_blessures, df_joueurs, left_on='ID', right_on='player_id', how='left')

# Fusionner les données résultantes avec les matchs en utilisant 'tourney_id'
df_complete = pd.merge(df_blessures_joueurs, df_matchs, left_on='tourney_id', right_on='tourney_id', how='left')

# Afficher un aperçu des données fusionnées
print(df_complete.head())
```

```python
print("Colonnes de df_blessures :")
print(df_blessures.columns)

print("\nColonnes de df_joueurs :")
print(df_joueurs.columns)

print("\nColonnes de df_matchs :")
print(df_matchs.columns)

print("\nColonnes de df_terrains :")
print(df_terrains.columns)
```
