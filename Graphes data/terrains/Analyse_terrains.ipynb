```python
import pandas as pd 
import matplotlib.pyplot as plt
```

I.Analyse df1 : tennis_courts

1.Lecture du dataset

```python
df1 = pd.read_csv('tennis_courtsV0.csv')
df1.head()
```

```python
df1.drop(columns=['clay', 'wall', 'grass', 'indoor', 'lights'], inplace=True)
df1.rename(columns={'clay.1': 'clay', 'wall.1': 'wall', 'grass.1': 'grass', 'indoor.1': 'indoor', 'lights.1': 'lights'}, inplace=True)
df1.head()
```

2.Name / city 

```python
terrain_par_ville = df1.groupby('city')['name'].count()
terrain_par_ville.plot(kind='bar', color='green')
plt.title('Nombre de terrains par ville')
plt.xlabel('Ville')
plt.ylabel('Nombre de terrains')
plt.xticks(rotation=45)  
plt.tight_layout() 
plt.show()
```
