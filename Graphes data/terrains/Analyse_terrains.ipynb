Ce notebook consiste à faire une analyse sur les terrains de tennis 

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

I.Analyse df1 : tennis_courts

1.Lecture du dataset

```python
df1 = pd.read_csv('tennis_courtsV1.csv')
df1.head()
```

```python
df1.drop(columns=['clay', 'wall', 'grass', 'indoor', 'lights'], inplace=True)
df1.rename(columns={'clay.1': 'clay', 'wall.1': 'wall', 'grass.1': 'grass', 'indoor.1': 'indoor', 'lights.1': 'lights'}, inplace=True)
df1.head()
```

2.Name / city 

```python
# Compter le nombre d'adresses répétées pour chaque valeur unique dans la colonne 'name'
address_counts = df1['name'].value_counts()

# Créer un graphique à secteurs
plt.figure(figsize=(10, 6))
plt.pie(address_counts, labels=address_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Répartition des adresses par établissement')
plt.axis('equal')  # Assure que le pie chart est un cercle
plt.tight_layout()
plt.show()
```
