```python
import pandas as pd 
import matplotlib.pyplot as plt
import missingno as msno
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tennis_court_nettoye.csv')
df.head()
```

```python
data = df[['count', 'clay', 'wall', 'grass', 'indoor', 'lights']]
totals = data.sum()
plt.figure(figsize=(8, 4))
bars = plt.bar(data.columns, data.sum())
for bar, total in zip(bars, totals):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.1, bar.get_height() - total * 0.5, f'{total}', ha='center', color='white')
plt.title('Répartition des Caractéristiques des Courts de Tennis')
plt.xlabel('Caractéristique')
plt.ylabel('Nombre de Courts')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
features = ['clay', 'wall', 'grass', 'indoor', 'lights']
df[features] = df[features].astype(int)
features_count = df[features].sum()
plt.figure(figsize=(7, 4))
features_count.plot(kind='bar')
plt.title('Présence de caractéristiques spécifiques dans les terrains de tennis')
plt.xlabel('Caractéristique')
plt.ylabel('Nombre de terrains')
plt.tight_layout()
plt.show()
```

```python
equip_cols = ['proshop', 'stringing']
equip_counts = df[equip_cols].sum()
plt.figure(figsize=(6, 4))
sns.barplot(x=equip_counts.index, y=equip_counts.values)
plt.title('Nombre de terrains de tennis avec équipements disponibles')
plt.xlabel('Équipement')
plt.ylabel('Nombre de terrains')
plt.show()
```

```python
df2 = pd.read_csv('gps-tennis-courts_nettoye.csv')
df2.head()
```

2. Diagramme circulaire pour la répartition des surfaces de jeu

```python
plt.figure(figsize=(4, 4))
surface_counts = df2['Surface'].value_counts()
plt.pie(surface_counts, labels=surface_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Répartition des surfaces de jeu')
plt.axis('equal')
plt.show()
```

 5. Graphique à barres pour l'accès public

```python
plt.figure(figsize=(7, 4))
sns.countplot(x='Acces_Public', data=df2)
plt.title('Nombre de terrains de tennis avec accès public')
plt.xlabel('Accès public')
plt.ylabel('Nombre de terrains')
plt.show()
```

Diagramme en violon pour la répartition des caractéristiques du court

```python
plt.figure(figsize=(8, 4))
sns.violinplot(x='Surface', y='Nombre_Terrains', data=df2)
plt.title('Répartition du nombre de terrains par surface de jeu')
plt.ylabel('Nombre de terrains')
plt.xlabel('Surface de jeu')
plt.show()
```
