Ce notebook vise à nettoyer et traiter les datasets des tournois des femmes wta 

```python
import pandas as pd 
import matplotlib.pyplot as plt
import missingno as msno
import matplotlib.pyplot as plt
import re
import json
import os
```

1.WTA (2007 - 2008 - 2009)

```python
df_2007 = pd.read_excel('wta_2007.xls')
df_2007.head()
```

```python
datasets_wta = ['wta_2007.xls', 'wta_2008.xls', 'wta_2009.xls']
dfs_wta = []
for dataset in datasets_wta:
    df = pd.read_excel(dataset)
    year = int(dataset.split('_')[1].split('.')[0])  
    df['year'] = year
    cols = ['year'] + [col for col in df.columns if col != 'year']
    df = df[cols]
    df = df.dropna(axis=1, how='all')
    df.fillna(0, inplace=True)
    msno.matrix(df)
    plt.title(f"Matrice de valeurs manquantes pour {dataset}")
    plt.show()
    dfs_wta.append(df)
    
    
```

```python
merged_df_wta = pd.concat(dfs_wta)
merged_df_wta.to_excel("wta_07_09.xlsx", index=False)
```

2. wta (2010 - 2024 )

```python
wta_datasets_xls = ['wta_2010.xls', 'wta_2011.xls', 'wta_2012.xls']

wta_datasets_xlsx = ['wta_2013.xlsx', 'wta_2014.xlsx', 'wta_2015.xlsx', 'wta_2016.xlsx', 'wta_2017.xlsx',
                     'wta_2018.xlsx', 'wta_2019.xlsx', 'wta_2020.xlsx', 'wta_2021.xlsx', 'wta_2022.xlsx',
                     'wta_2023.xlsx', 'wta_2024.xlsx']
dfs_wta = []
for dataset in wta_datasets_xls:
    df = pd.read_excel(dataset)
    year = int(dataset.split('_')[1].split('.')[0])
    df['year'] = year
    cols = ['year'] + [col for col in df.columns if col != 'year']
    df = df[cols]
    df = df.dropna(axis=1, how='all')
    df.fillna(0, inplace=True)
    dfs_wta.append(df)
for dataset in wta_datasets_xlsx:
    df = pd.read_excel(dataset)
    year = int(dataset.split('_')[1].split('.')[0])
    df['year'] = year
    cols = ['year'] + [col for col in df.columns if col != 'year']
    df = df[cols]
    df = df.dropna(axis=1, how='all')
    df.fillna(0, inplace=True)
    dfs_wta.append(df)
merged_df_wta = pd.concat(dfs_wta)
merged_df_wta.to_excel("merged_wta_datasets.xlsx", index=False)
```
